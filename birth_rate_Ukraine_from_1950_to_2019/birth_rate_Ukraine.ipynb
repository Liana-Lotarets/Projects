# Опис дослідження

Досліджується таблиця "Коефіцієнт народжуваності в регіонах України (1950—2019)" з сайту https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

__Таблиця "Коефіцієнт народжуваності в регіонах України (1950—2019)"__

```python
URL = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"
table = pd.read_html(
    URL,
    match="Коефіцієнт народжуваності в регіонах України",
    thousands=".",
    decimal=",",
)[0]
```

```python
table
```

Кількість рядків та стовпців таблиці.

```python
rows_and_columns_number = table.shape
print(f"{rows_and_columns_number[0]} рядків, {rows_and_columns_number[1]} стовпців.")
```

Зробимо заміну значення "—" на значення NaN.

```python
table.replace("—", np.nan, inplace=True)
```

```python
table
```

Типи всіх стовпців.

```python
table.dtypes
```

Зробимо заміну типів нечислових колонок на числові. 

```python
columns_years_list = table.columns[1:]
table[columns_years_list] = table[columns_years_list].apply(pd.to_numeric)
print(table.dtypes)
```

Кількість пропусків, що міститься в кожній колонці.

```python
table.isnull().sum()
```

Видалимо з таблиці останній рядок (дані по всій країні).

```python
table = table.drop([rows_and_columns_number[0] - 1])
```

```python
table
```

Зробимо заміну відсутніх даних в стовпцях середніми значеннями цих стовпців.

```python
table[columns_years_list] = (
    table[columns_years_list].fillna(table[columns_years_list].mean()).round(1)
)
```

```python
table
```

__Середнє значення коефіцієнту народжуваності по Україні__

```python
mean_years_array = np.array(table[columns_years_list].mean())
mean_years_table = pd.DataFrame({'Mean': mean_years_array}, index=columns_years_list)
```

```python
mean_years_table
```

```python
mean_years_table.plot(
    kind="area", title="Середнє значення коефіцієнту народжуваності по Україні", xlabel="Рік", ylabel="Коефіцієнт народжуваності",
)
plt.show()
```

__Висновок__

Локально народжуваність наприкінці 1990-х — на початку 2000-х була найнижчою за період 1950–2019 років.

Деталізуємо інформацію по регіонах.

__Коефіцієнт народжуваності у 2000 році по регіонах__

```python
table_2000 = table[["Регіон", "2000"]].sort_values(by="2000").reset_index(drop=True)
table_2000.columns = ["Регіон", "Коефіцієнт"]
```

```python
table_2000
```

```python
table_2000.plot(
    kind="barh", x="Регіон", y="Коефіцієнт", xlabel="Коефіцієнт народжуваності"
)
plt.show()
```

Перший та третій квартілі.

```python
first_quartile, third_quartile = (
    table_2000["Коефіцієнт"].quantile([0.25, 0.75]).to_list()
)
print(f"Перший квартіль: {first_quartile}\nТретій квартіль: {third_quartile}")
```

__Список регіонів, де коефіцієнт народжуваності у 2000 році був вищим за третій квартіль__

```python
table_more_third_quartile = table_2000[table_2000["Коефіцієнт"] > third_quartile][
    ["Регіон", "Коефіцієнт"]
].sort_values(by="Коефіцієнт", ascending=False)
```

```python
table_more_third_quartile
```

__Висновок__

Найвища народжуваність у 2000 році була у Рівненській та Закарпатській областях. Далі йдуть Волинська, 	Івано-Франківська, Чернівецька, Тернопільська та Львівська області.

__Список регіонів, де коефіцієнт народжуваності у 2000 році був нижчим за перший квартіль__

```python
table_less_first_quartile = table_2000[table_2000["Коефіцієнт"] < first_quartile][
    ["Регіон", "Коефіцієнт"]
].sort_values(by="Коефіцієнт")
```

```python
table_less_first_quartile
```

__Висновок__

Найнижча народжуваність у 2000 році була у Донецькій та Луганській областях. Далі йдуть Харківська, Чернігівська, Сумська, Полтавська області та Севастополь.
